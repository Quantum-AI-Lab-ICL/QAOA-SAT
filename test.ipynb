```python
from qiskit.visualization import plot_histogram
from max_3_sat.quantum_solver import QuantumSolver
from formula.variable import Variable
from formula.clause import Clause
from formula.wcnf import WCNF

x0 = Variable(0, False) 
x1 = Variable(1, False) 
x2 = Variable(2, False) 
notx0 = Variable(0, True) 
notx1 = Variable(1, True) 
notx2 = Variable(2, True) 

c1 = Clause([x0, x1, x2])
c2 = Clause([x0, x1, notx2])
c3 = Clause([x0, notx1, x2])
c4 = Clause([x0, notx1, notx2])
c5 = Clause([notx0, x1, x2])
c6 = Clause([notx0, x1, notx2])
c7 = Clause([notx0, notx1, x2])

wcnf = WCNF([c1, c2, c3, c4, c5, c6, c7])
print(wcnf)
for ass in [a + b + c for a in "01" for b in "01" for c in "01"]:
  print(f'is satisfied by {ass}: {wcnf.is_satisfied(ass)}')
```

```python
quantum_solver = QuantumSolver()
quantum_solver.max_sat(wcnf, ret_num=1)
```

```python
plot_histogram(quantum_solver.result)
```
